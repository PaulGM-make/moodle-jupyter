## 🔹 Variables y Tipos de Datos

Una variable es un espacio en memoria donde almacenamos un valor. En Python, no es necesario declarar el tipo, ya que es dinámico.

Ejemplos comunes en electrónica:
- Voltajes
- Estados lógicos
- Nombres de sensores


````python
# Declaración de variables

voltaje = 3.3       # voltaje en voltios
corriente = 0.015   # corriente en amperios
activo = True       # sensor activado (estado booleano)
sensor = "LDR"      # tipo de sensor (Light Dependent Resistor)

print("Voltaje:", voltaje)
print("Corriente:", corriente)
print("¿Está activo?", activo)
print("Sensor:", sensor)
````

## 🔁 Estructuras Condicionales

Las estructuras de control nos permiten tomar decisiones. En aplicaciones electrónicas esto se usa, por ejemplo, para activar un ventilador si la temperatura es mayor a cierto valor.


````python
temperatura = 28  # grados Celsius

if temperatura > 40:
    print("🔥 Alerta: sobrecalentamiento")
elif temperatura > 30:
    print("⚠️ Temperatura elevada")
else:
    print("✅ Temperatura normal")
````

## 🔁 Bucles: for y while

Los bucles nos permiten repetir instrucciones. Por ejemplo, para tomar lecturas de un sensor cada segundo o analizar múltiples valores registrados.


````python
# Bucle for: recorrer una lista de voltajes
voltajes = [3.3, 3.1, 3.0, 2.9]

for v in voltajes:
    print("Medición:", v, "V")
````

````python
# Bucle while: simular lectura hasta que el voltaje sea seguro

voltaje = 2.5

while voltaje < 3.0:
    print("⚠️ Voltaje bajo:", voltaje)
    voltaje += 0.2  # simulamos subida de voltaje

print("✅ Voltaje estable:", voltaje)
````

## 🧰 Listas

Las listas se usan para almacenar múltiples datos (como muestras de un sensor).

### ✅ Características principales:
- Se accede a los elementos por su **índice**, comenzando desde 0.
- Se puede **modificar**, **agregar** y **eliminar** elementos.
- Se pueden realizar operaciones como **recorrer** la lista con bucles, ordenarla, invertirla, etc.

### Métodos útiles con listas

- `.append(x)` → Agrega un elemento al final.
- `.insert(i, x)` → Inserta un elemento en una posición específica.
- `.remove(x)` → Elimina el primer elemento con el valor `x`.
- `.pop()` → Elimina el último elemento.
- `.sort()` → Ordena los elementos.
- `.reverse()` → Invierte el orden.

Ejemplo:
```python
mi_lista = [3, 1, 4, 2]
mi_lista.sort()
print(mi_lista)  # [1, 2, 3, 4]
```


````python
# Lista con temperaturas en °C
temperaturas = [21.5, 22.0, 22.3, 21.9, 23.1]

# Mostrar solo las temperaturas mayores a 22
for t in temperaturas:
    if t > 22:
        print("🌡️ Alta:", t)
````

````python
# Crear y manipular una lista
numeros = [10, 20, 30, 40, 50]
print("Lista original:", numeros)

# Acceder al primer y último elemento
print("Primer elemento:", numeros[0])
print("Último elemento:", numeros[-1])

# Agregar un nuevo número al final
numeros.append(60)
print("Lista modificada:", numeros)

# Eliminar un elemento (por índice)
del numeros[1]
print("Lista tras eliminar el segundo elemento:", numeros)
````

## 🧮 Diccionarios en Python

Un **diccionario** es una colección no ordenada de pares **clave: valor**. Se utilizan para almacenar información estructurada, como en un registro o ficha.

Se definen con llaves `{}`, y cada elemento está compuesto por una clave única y un valor.

### ✅ Características principales:
- Se accede a los valores usando claves.
- Las claves pueden ser de tipo string, int, etc.
- Es mutable: se pueden agregar, modificar y eliminar pares clave-valor.

### 🧠 Ejemplo básico:
```python
persona = {
    'nombre': 'Lucía',
    'edad': 25,
    'ciudad': 'Cuenca'
}
print(persona['nombre'])  # Lucía
```

### Métodos útiles con diccionarios

- `.keys()` → Devuelve todas las claves.
- `.values()` → Devuelve todos los valores.
- `.items()` → Devuelve pares clave-valor.
- `.get(clave)` → Devuelve el valor de una clave, si existe.
- `.pop(clave)` → Elimina el par con esa clave.

Ejemplo:

```python
estudiante = {'nombre': 'Pedro', 'nota': 9.5}
print(estudiante.get('nota'))  # 9.5
```

````python
# Crear y manipular un diccionario
persona = {
    'nombre': 'Carlos',
    'edad': 28,
    'ciudad': 'Guayaquil'
}

# Acceder a un valor
print("Nombre:", persona['nombre'])

# Agregar un nuevo par clave-valor
persona['profesion'] = 'Ingeniero'

# Modificar un valor
persona['edad'] = 29

# Eliminar una clave
del persona['ciudad']

# Mostrar el resultado
print("Diccionario actualizado:", persona)
````

## 🧮 Funciones

Las funciones permiten reutilizar bloques de código. Por ejemplo, podríamos crear una función que calcule la potencia (P = V * I) cada vez que recibimos una lectura.


````python
# Definición de función
def calcular_potencia(voltaje, corriente):
    potencia = voltaje * corriente
    return potencia

# Uso de la función
v = 5.0
i = 0.02
p = calcular_potencia(v, i)
print(f"🔋 Potencia: {p} W")
````
